```python
#import numpy, pandas, dan scikit-learn
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
#membaca dataset dr file ke pandas dataFrame
irisDataset = pd.read_csv('klasifikasi_dataset_iris.csv', delimiter=';', header=0)
```

```python
irisDataset.head()
```

```python
#mengubah kelas (kolom "species") dari string ke unique-integer
irisDataset['Species'] = pd.factorize(irisDataset.Species)[0]
```

```python
irisDataset.head()
```

```python
print(irisDataset)
```

```python
#menghapus kolom "Id"
irisDataset = irisDataset.drop(labels="Id", axis=1)
```

```python
print(irisDataset)
```

```python
#mengubah dataFrame ke array Numpy
irisDataset = irisDataset.to_numpy()
```

```python
print(irisDataset)
```

```python
#membagi dataset 160 baris utk data training dan 40 baris utk data testing
dataTraining = np.concatenate((irisDataset[0:80, :], irisDataset[100:180, :]), axis=0)
dataTesting = np.concatenate((irisDataset[80:100, :], irisDataset[180:200, :]), axis=0)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
#memecah dataset ke input dan label
inputTraining = dataTraining[:, 0:4]
inputTesting = dataTesting[:, 0:4]
labelTraining = dataTraining[:, 4]
labelTesting = dataTesting[:, 4]
print(labelTraining)
len(labelTraining)
```

```python
#mendefinisikan decision tree classifier 
model = tree.DecisionTreeClassifier()
```

```python
#mentraining model
model = model.fit(inputTraining, labelTraining)
```

```python
#memprediksi input data testing
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
#menghitung akurasi 
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar : ", prediksiBenar, "data")
print("Prediksi Salah : ", prediksiSalah, "data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah)*100,"%")
```
